# Working with more data

```python
confirmed = pd.read_csv("data/time_series_covid19_confirmed_global.csv")
death = pd.read_csv("data/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("data/time_series_covid19_recovered_global.csv")
```

```python
confirmed
```

```python
# get the last date from the dataset. More recent data.
def latest_by_country(data):
    return data.groupby("Country/Region").sum(numeric_only=True).iloc[:, -1]
```

```python
cases = [latest_by_country(confirmed), latest_by_country(death), latest_by_country(recovered)]

df_cases = pd.concat(cases, axis=1)
```

```python
df_cases.columns = ["confirmed", "death", "recovered"]
df_cases.head()
```

```python
sum_up = df_cases.sum()
lethality_rate_1 = sum_up["death"] / sum_up["confirmed"]*100
```

```python
print(lethality_rate_1)
```

The lethality represents the number of deaths divided by the number of confirmed cases. The lethality rate is a very important metric to understand the impact of the virus in a country. The higher the lethality rate, the more dangerous the virus is.

```python
# lethality rate by country
lethality_rate_1 = df_cases["death"] / df_cases["confirmed"]*100
```

```python
print(lethality_rate_1)
```

```python
# create a new column in df_cases with lethality rate_1
df_cases["lethality_rate_1"] = lethality_rate_1
df_cases.head()
```

```python
df_cases.sort_values("lethality_rate_1", ascending=False).head(10)
```

# Visualizing

```python
import seaborn as sns
```

```python
sns.scatterplot(x="confirmed", y="lethality_rate_1", data=df_cases)
```

# Adding dates

```python
def latest_by_country_at(data, date):
    return data.groupby("Country/Region").sum(numeric_only=True)[date]
```

```python
cases = [latest_by_country_at(confirmed,'3/9/21'), latest_by_country_at(death,'3/9/21'), latest_by_country_at(recovered,'3/9/21')]

df_cases_3_9_21 = pd.concat(cases, axis=1)
df_cases_3_9_21.columns = ["confirmed", "death", "recovered"]
df_cases_3_9_21.head()
```

```python
sum_up = df_cases_3_9_21.sum()
lethality_rate_1 = sum_up["death"] / sum_up["confirmed"]*100
lethality_rate_2 = sum_up["death"] / (sum_up["death"] + sum_up["recovered"])*100
print(lethality_rate_1)
print(lethality_rate_2)
```

```python
# check the lethality rate in Brazil
sum_up = df_cases_3_9_21.loc["Brazil"]
lethality_rate_1 = sum_up["death"] / sum_up["confirmed"]*100
lethality_rate_2 = sum_up["death"] / (sum_up["death"] + sum_up["recovered"])*100
print(lethality_rate_1)
print(lethality_rate_2)
```

```python
# check the lethality rate in China
sum_up = df_cases_3_9_21.loc["China"]
lethality_rate_1 = sum_up["death"] / sum_up["confirmed"]*100
lethality_rate_2 = sum_up["death"] / (sum_up["death"] + sum_up["recovered"])*100
print(lethality_rate_1)
print(lethality_rate_2)
```
